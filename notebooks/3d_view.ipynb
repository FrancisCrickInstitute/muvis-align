# 3d view using napari

```python
import sys
sys.path.append('..')

import napari
import numpy as np

from muvis_align.image.source_helper import create_dask_source
from muvis_align.image.util import invert_data
from muvis_align.util import get_filetitle, print_hbytes
```

## Load data

```python
filename = 'D:/slides/EM04652-02_slice17_spaghettiandmeatballs2/aligned_ov_hpc/registered.ome.zarr'
min_level = 2   # choose lowest resolution level to load, to force napari rendering
do_invert = True

source = create_dask_source(filename)
data = source.get_data(level=-1)[:min_level]

if do_invert:
    data = invert_data(data)

name = get_filetitle(filename)
scale = source.get_pixel_size(asarray=True)
translation = source.get_position(asarray=True)
if len(scale) == 0:
    scale = None
else:
    print('scale:', source.get_pixel_size())
if len(translation) == 0:
    translation = None

print('shape')
for d in data:
    print(d.shape, print_hbytes(np.prod(d.shape) * d.itemsize))
```

## Show image in napari

```python
viewer = napari.Viewer(ndisplay=3)
viewer.add_image(data, name=name, scale=scale, translate=translation, rendering='average', metadata=source.metadata)
```
